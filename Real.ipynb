```python
# Risk & Return Analysis
col
```

```python
adj_close_df
```

```python
for i in col:
    adj_close_df[i+'%Return'] = adj_close_df[i].pct_change()
```

```python
adj_close_df.dropna(inplace=True)
```

```python
return_columns = list(adj_close_df.columns)[-10:]

return1 = []
risk1 = []

for i in return_columns:
    temp = adj_close_df[i].mean()
    temp2 = ((temp + 1)**252 - 1) * 100
    return1.append(temp2)
    
    temp3 = adj_close_df[i].std() * np.sqrt(252) * 100
    risk1.append(temp3)
```

```python
risk_return_df = pd.DataFrame(index=col)
risk_return_df['Risk(%)'] = risk1
risk_return_df['Return(%)'] = return1
risk_return_df['Risk Adj. Return'] = (risk_return_df['Return(%)'] / risk_return_df['Risk(%)'])
```

```python
risk_return_df
```

```python
risk_return_df.to_csv('Risk Adj Return.csv')
```

```python
risk_return_list = list(risk_return_df['Risk Adj. Return'])
```

```python
sum_list = np.sum(risk_return_list)
```

```python
weight_array = risk_return_list / sum_list
```

```python
weight_array.sum()
```

```python
risk_return_df['weightage'] = weight_array
```

```python
risk_return_df
```

```python
return_df = adj_close_df.iloc[:,10:]
```

```python
return_df.columns
```

```python
rebased_df = pd.DataFrame()

for j in list(return_df.columns):
    rebase_price_list = [100]
    for i in list(return_df[j]):
        temp = rebase_price_list[-1] * (i+1)
        rebase_price_list.append(temp)
    rebased_df[j] = rebase_price_list
    
```

```python
rebased_df
```

```python
plt.figure(figsize=(20,9))
plt.plot(rebased_df)
plt.legend(rebased_df.columns)
plt.show()
```

```python
rebased_df.dot(list(risk_return_df['weightage']))
```

```python
final_df = pd.DataFrame()
final_df['portfolio NAV'] = rebased_df.dot(list(risk_return_df['weightage']))
```

```python
ticker = 'SPY'

spy = yf.download(ticker, period='10y', interval='1d',auto_adjust=False)
spy['SPY %Return'] = spy['Adj Close'] / spy['Adj Close'].shift(1) - 1
spy.dropna(inplace=True)

rebased_price_list = [100]
for i in list(spy['SPY %Return']):
    temp = rebased_price_list[-1] * (i + 1)
    rebased_price_list.append(temp)
```

```python
final_df['S&P500'] = rebased_price_list
```
